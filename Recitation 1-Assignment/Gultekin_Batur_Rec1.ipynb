##Question 2

Write a function nums_reversed that takes in an integer n and returns a string containing the numbers 1 through n including n in reverse order, separated by spaces. For example:

`>>>nums_reversed(5)` 

`'5 4 3 2 1' `

```
# Your code 

def nums_reversed (intnumber):
  if intnumber <= 0:
    print("Wrong type of input")
  else:
    try:
      st=""
      for i in range (intnumber):
        st= st + " " + str(intnumber-i)
      st = st[1:]  
      return st
    except:
        print("Wrong type of input") 
        
nums_reversed(5)
```

##Question 3

Write a function doubleN that takes in an integer N and a list of integers and returns True only if the list has two instances of the integer N next to each other.

For example:

`>>> double100(100, [100, 2, 3, 100])`

`False`

`>>> double100(100, [2, 3, 100, 100, 5])`

`True`

```
# Your code

def doubleN(N, listofInt):
  for i in range(len(listofInt)-1):
    if N==listofInt[i] and N==listofInt[i+1]:
      return True
  return False

doubleN(100,[100, 100, 3, 100])
  
```
